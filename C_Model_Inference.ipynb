```python
import pandas as pd
import joblib

# Cargar el modelo entrenado
model = joblib.load('house_prices_model.pkl')

# Cargar datos de prueba (si es necesario)
X_new = pd.read_csv('data/processed_X.csv')  # Usa datos nuevos si lo deseas
```

```python
# Hacer predicciones con los datos cargados
y_pred_new = model.predict(X_new)

# Mostrar las primeras 10 predicciones
print("Predicciones:", y_pred_new[:10])
```

```python
# Guardar las predicciones en un archivo CSV
pd.DataFrame(y_pred_new, columns=['Predicciones']).to_csv('data/predicciones.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Visualizar algunas de las predicciones
plt.figure(figsize=(10, 6))
plt.plot(y_pred_new[:50], label='Predicciones', marker='o')
plt.title('Predicciones de los primeros 50 datos')
plt.legend()
plt.show()
```
